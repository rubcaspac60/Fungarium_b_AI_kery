# Descrition of Physical properties
The physico-mechanical properties of the biocomposites are calculated using the following functions and are mostly obtained with standardized machinery. We acknowledge that some of these value could be calculated using other machines or methods than stipulated. We give complete freedom to our partners in choosing the most suitable method for obtaining the values with the condition of keeping consistency and accuracy. We would like to keep a standard deviation of our values at $<5\%$ with a minima of 5 test runs for averaging.

##Dry density
It is calculated by taking the ratio of the dried mass over the selected volume. 

$$ \sigma = \frac{m_{dry}}{V_{standardized}} $$

where, m is mass (g) and V is volume ($mm^3$)

##Moisture Content
It is the difference of the humid and dry sample weight and transformed into a percentage by dividing by the mass of dry sample. 

$$ M = \frac{(W_h - w_d)*100}{W_d} $$
where: M = Moisture content, $W_h$ = humid weight (g, Kg), $W_d$ = dry weight (g, Kg)

## Mechanical Testing in Compression
These is done in a Instron 5900R lead bench with a 100 kN capacity and a 10 kN load cell at RT. Or a Zwick/Roell Z010 universal testing machine using an elongation rate of 2 mm/min and a maximum force of 1 kN.   The stop sign of the experiment if when a strain was reached at 70 - 80 \% 
The obtained load-displacement curve is then converted to a stress-strain curve, using:

$$ \sigma = \frac{F}{A} $$
$$ \epsilon = \frac{\delta L}{L_o} $$

where, $\sigma$ is the compressive stress, $\epsilon$ is the compressive strain, F is the compressive force (N), A is the original cross-section of the sample ($mm^2$), $\delta L$ is the displacement of the loading surfaces (mm) and $L_o$ is the original height of the test piece (mm). 

Flexural tests are  performed in a three-point bending setup with the same machine using a cross-head speed of 2 mm/min and clamp support distance of 80 mm.
Data is analysed to obtain stress-strain plots, tensile ($St_{ten}$) and flexural strength ($St_{flex}$), and the elastic ($M_{ela}$) and flexural modulus ($M_{flex}$), Young modulus ($M_{young}$), Elongation at break ($Bk_{elo}$). The test displacement is controlled at a rate of 2 mm/min.

## Thermal Conductivity
Several methods could be used to obtain a correlation of this property. A dynamical methods of the insertion of a thermal needle probe (TNP) which complies with the standard test methods for determination of thermal conductivity of soil and soft rock needle probe procedure (ASTM D 5334-00). A first guide hole (2mm) was drilled in the centre of the cylindrical sample with the help of an extra needle with a smaller diameter than the TNP, a different electrical current in the circuit was applied. When a high current is applied, the temperature increases faster, resulting in a lower thermal conductivity. The fast rise in temperature is not recommended by the standard due to possible errors while reading. In order to have reliable results, an average was taken between high currents and low currents. The thermal conductivity is determined by:

$$ \lambda = \frac{Q}{4\pi (T_2 - T_1)}* ln\frac{t_2}{t_1} $$

where Q is constant current ($\frac{W}{m}$), $T_n$ is the temperature (C and K), and $t_n$ is the time of linear proportion (s). 

## Electrical Four Point Conductivity
Four-point probe measurer (FPP) is a tool that is commonly used to measure resistivity value of a layer of an electronic material. Like its name, this measurement relies on 4 probes; 2 probes are for electric current and the 2 other ones are to measure voltage when they are applied to the material. To determine and to review the characteristics of the material, the resistivity value need to be measured for certain area and thickness. Some other parameters that can be obtained from material measurement using this tool is to know the type of doping of a semi-conductor material (positive and negative) and electronic mobility of a material.

It is initially assumed that the tip of the probe is very small (infinitesimal) and the sample has a large dimension (semi-infinite). This bulk sample requires that the sample thickness (t) be more than the distance between probes (s).

$$ res = 2\pi s \frac{V}{I} $$
with :
$$ C_v = \frac{1}{res} $$

## Rate of Water Absorption
The method for Measurement of Rate of Absorption of Water by Hydraulic-Cement Concretes (ASTM C 1585) with partially submerged samples was applied to the mycelium composites. The tests were conducted at RT. The water absorption was determined from the weight difference in relation to the initial weight. The formula to determine the water absorption is:

$$ m = \frac{M_t}{a*d} $$

where, m is the water absorption (mm), $M_t$ is the changed weight pf the specimen (g, kg), a is the exposed cross-section area ($mm^2$) and d is the density of the water ($\frac{g}{mm^3}$).
